```python
import torch 
import torch.nn as nn 
##tiene el tamaño de los tensores de entradas y le tamaños de los tensores de salida 
#capa lineal
linear = nn.Linear(in_features=4096,out_features=10) #solo toma tensores que son de dimension 1 input
##capa convulocional en este caso usaremos una capa 2d que son para imagenes y 3d para videos 
# 
conv = nn.Conv2d(in_channels = 3,out_channels = 9,kernel_size = 3,stride = 1,padding = 1) #modificara los canales pero no el tamaño de la imagen 
relu = nn.ReLU(False)
```

```python
x = torch.randn(4096)
y = linear(x)
y.size()
```
